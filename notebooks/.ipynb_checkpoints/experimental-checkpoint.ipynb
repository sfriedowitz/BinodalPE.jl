## Data Load

```julia
data_exp = "data/exp"
img_exp = "images/exp"

if !isdir(data_exp); mkdir(data_exp) end
if !isdir(img_exp); mkdir(img_exp) end
    
;
```

```julia
exp_names = ["1A1C", "2A1C", "3A1C", "1A3C"]
exp_labels = ["1:1 A:C", "2:1 A:C", "3:1 A:C", "1:3 A:C"]

exp_bndls = []
for name in exp_names
    bndl_name = joinpath(data_exp, "exp_$(name).json")
    b = readbndl(bndl_name)
    push!(exp_bndls, b)
end

exp1A1C = exp_bndls[1]
exp2A1C = exp_bndls[2]
exp3A1C = exp_bndls[3]
exp1A3C = exp_bndls[4]

;
```

```julia
exp_pa_errs = []
exp_pc_errs = []
for bndl in exp_bndls
    err_pa = []
    err_pc = []
    for i = 1:length(bndl)
        state = bndl[i]
        push!(err_pa, state.props[:ea_sup])
        push!(err_pc, state.props[:ec_sup])
    end
    for i = length(bndl):-1:1
        state = bndl[i]
        push!(err_pa, state.props[:ea_coac])
        push!(err_pc, state.props[:ec_coac])
    end
    
    push!(exp_pa_errs, err_pa)
    push!(exp_pc_errs, err_pc)
end
exp_pol_errs = [ea + ec for (ea, ec) in zip(exp_pa_errs, exp_pc_errs)]

;
```

```julia
# Load peptide binodals
peptide_names = ["1A1C", "3A1C"]
peptide_labels = ["1:1 PGA:PLL", "3:1 PGA:PLL"]

peptide_bndls = []
for name in peptide_names
    bndl_name = joinpath(data_exp, "exp_peptide_$(name).json")
    b = readbndl(bndl_name)
    push!(peptide_bndls, b)
end

peptide1A1C = peptide_bndls[1]
peptide3A1C = peptide_bndls[2]

pepa_errs = []
pepc_errs = []
for bndl in peptide_bndls
    pa_err = []
    pc_err = []
    for i = 1:length(bndl)
        state = bndl[i]
        push!(pa_err, state.props[:ea_sup])
        push!(pc_err, state.props[:ec_sup])
    end
    for i = length(bndl):-1:1
        state = bndl[i]
        push!(pa_err, state.props[:ea_coac])
        push!(pc_err, state.props[:ec_coac])
    end
    
    push!(pepa_errs, pa_err)
    push!(pepc_errs, pc_err)
end

# Load turbidity data
exp_turbidity = readdlm(joinpath(data_exp, "exp_turbidity.csv"), ',')

;
```

## Synthetic PEs

### 1:1 A:C

```julia
raw_xlsx = XLSX.readxlsx(joinpath(data_exp, "raw_1A1C_N50.xlsx"))
sheet = raw_xlsx["Sheet1"]
data = sheet["A1:P32"]

ca_bulk, cc_bulk = 0.05, 0.05
cs_bulk = data[3:9,1]

vol_sup = data[3:9,2]
vol_coac = data[3:9,6]

# Polycation
cc_sup = data[3:9,4] * 1e-3
ec_sup = data[3:9,5] * 1e-3

cc_coac = data[3:9,8] * 1e-3
ec_coac = data[3:9,9] * 1e-3

# Polyanion
ca_sup = data[14:20,4] * 1e-3
ea_sup = data[14:20,5] * 1e-3

ca_coac = data[14:20,8] * 1e-3
ea_coac = data[14:20,9] * 1e-3

# Cation
cp_sup = data[14:20,11]
ep_sup = data[14:20,12]

cp_coac = data[14:20,13]
ep_coac = data[14:20,14]

# Anion
cm_sup = data[3:9,11]
em_sup = data[3:9,12]

cm_coac = data[3:9,13]
em_coac = data[3:9,14]

;
```

```julia
data_sup = convert(Array{Float64,2}, hcat(ca_sup, cc_sup, cp_sup, cm_sup))
data_coac = convert(Array{Float64,2}, hcat(ca_coac, cc_coac, cp_coac, cm_coac))

data_bulk = similar(data_sup)
for (i, row) in enumerate(eachrow(data_bulk))
    row[1] = ca_bulk
    row[2] = cc_bulk
    row[3] = cs_bulk[i] + ca_bulk
    row[4] = cs_bulk[i] + cc_bulk
end

exp1A1C = BinodalData()
for i = 1:size(data_bulk,1)
    state = BinodalState(data_bulk[i,:])
    state.sup = data_sup[i,:]
    state.dense = data_coac[i,:]
    state.nu = vol_coac[i] / (vol_sup[i] + vol_coac[i])
    
    # Error bars
    state.props[:cs_bulk] = cs_bulk[i]
    state.props[:ea_sup] = ea_sup[i]
    state.props[:ea_coac] = ea_coac[i]
    
    state.props[:ec_sup] = ec_sup[i]
    state.props[:ec_coac] = ec_coac[i]
    
    state.props[:ep_sup] = ep_sup[i]
    state.props[:ep_coac] = ep_coac[i]
    
    state.props[:em_sup] = em_sup[i]
    state.props[:em_coac] = em_coac[i]
    
    add_state!(exp1A1C, state)
end

savebndl(exp1A1C, joinpath(data_exp, "exp_1A1C.json"))
```

# Theory Fitting

**Design rules for looping + association model:**

- Coil-like electrostatics lead to looping-back at low salt
- Can be obtained with coil structure, and low binding energies
- Low binding energy --> higher charge density, more looping

## Weak Binding

```julia
wP = 10
wS = 3
np = 50
chi = 0.5
dg = [-1.0, -1.0, -1.55]
b = 0.75 * wP^(1/3)

model = AssociationCoacervate(structure = GaussianCoil,
    omega = [wP, wP, wS, wS], dg = dg, chi = [chi, chi, 0.0],
    np = [np, np], b = [b, b]
)

;
```

### s = 0.0 (1:1 A:C)

```julia
concAB, concCB = asypolyion(0.1, 0.0)
phiAB, phiCB = tophi(concAB, model.omega[1]), tophi(concCB, model.omega[2])

(phiAB, phiCB)
```

```julia
salt = collect(range(0.062, 0.0; length = 50))
bulk = neutralbulk([phiAB, phiCB, salt[1]], model)
set_bulk!(model, bulk)

#init = [0.002384598206929761, 0.0023845982069301327, 0.06501440118335282, 0.06501440118335292, 0.021639084100907105, 0.021639084100906227, 0.06395522225436542, 0.06395522225436513, 0.29683481680796125, -1.4589164433983702e-14]
#res = bndlsolve(init, model; verbose = true, scale = true, iter = 25)

init = [0.02956259480268428, 0.029496835501715928, 0.8149943050191757, 0.1136278856384763]
res = bndlminimize(init, model; verbose = true, scale = false, iter = 250, cycles = 5)
```

```julia
init = res.state
bndl1A1C = BinodalData()
for phiS in salt
    println("Working on phiS = $(phiS)")
    
    bulk = bulk = neutralbulk([phiAB, phiCB, phiS], model)
    set_bulk!(model, bulk)
    
    res = bndlsolve(init, model; verbose = false, scale = true, iter = 50, tol = 1e-8)
    
    if res.converged
        state = res.state
        state.props[:cs_bulk] = toconc(phiS, wS)
        add_state!(bndl1A1C, state)    
    
        init = state
    else
        break
    end
end
```

```julia
sort!(bndl1A1C, 3)

fig = plot(bndl1A1C,
    map = x -> [x[1]+x[2], x[3]+x[4]],
    tie = true,
    composition = :conc,
    omega = model.omega
) 

plot!(fig, exp1A1C,
    map = x -> [x[1]+x[2], x[3]+x[4]],
    marker = :diamond,
    tie = false
)
```

```julia
fig = plot(bndl1A1C,
    map = x -> [x[1], x[3]+x[4]],
    composition = :conc,
    omega = model.omega
)

plot!(fig, exp1A1C, map = x -> [x[1], x[3]+x[4]], primary = false, la = 0.25, marker = :diamond)

plot!(fig, bndl1A1C,
    map = x -> [x[2], x[3]+x[4]],
    composition = :conc,
    omega = model.omega
)

plot!(fig, exp1A1C, map = x -> [x[2], x[3]+x[4]], primary = false, la = 0.25, marker = :diamond)
```

## Strong Binding

```julia
wP = 10
wS = 3
np = 50
chi = 0.5
dg = [-4, -4, -7]
b = 1.0 * wP^(1/3)

model = AssociationCoacervate(structure = GaussianCoil,
    omega = [wP, wP, wS, wS], dg = dg, chi = [chi, chi, 0.0],
    np = [np, np], b = [b, b]
)

;
```

### s = 0.0 (1:1 A:C)

```julia
concAB, concCB = asypolyion(0.1, 0.0)
phiAB, phiCB = tophi(concAB, model.omega[1]), tophi(concCB, model.omega[2])

(phiAB, phiCB)
```

```julia
salt = collect(range(0.0615, 0.0; length = 50))
bulk = neutralbulk([phiAB, phiCB, salt[1]], model)
set_bulk!(model, bulk)

#init = [0.002384598206929761, 0.0023845982069301327, 0.06501440118335282, 0.06501440118335292, 0.021639084100907105, 0.021639084100906227, 0.06395522225436542, 0.06395522225436513, 0.29683481680796125, -1.4589164433983702e-14]
#res = bndlsolve(init, model; verbose = true, scale = true, iter = 25)

init = [0.04312996566112919, 0.04307070135374988, 0.7900830949479909, 0.10509913411308572]
res = bndlminimize(init, model; verbose = true, scale = false, iter = 250, cycles = 5)
```

```julia
init = res.state
bndl1A1C = BinodalData()
for phiS in salt
    println("Working on phiS = $(phiS)")
    
    bulk = bulk = neutralbulk([phiAB, phiCB, phiS], model)
    set_bulk!(model, bulk)
    
    res = bndlsolve(init, model; verbose = false, scale = true, iter = 50, tol = 1e-8)
    
    if res.converged
        state = res.state
        state.props[:cs_bulk] = toconc(phiS, wS)
        add_state!(bndl1A1C, state)    
    
        init = state
    else
        break
    end
end
```

```julia
sort!(bndl1A1C, 3)

fig = plot(bndl1A1C,
    map = x -> [x[1]+x[2], x[3]+x[4]],
    tie = true,
    composition = :conc,
    omega = model.omega
) 

plot!(fig, exp1A1C,
    map = x -> [x[1]+x[2], x[3]+x[4]],
    marker = :diamond,
    tie = false
)
```
